```python
#Goals:
#1 - I will analyze which sex in this dataset tends to smoke more: men or women. 
#2 - I will also analyze which sex tends to gain more and less BMI.
#3 - for men and women who are smokers, where the majority of individuals are from.
#4 - The average insurance cost for men and for women in this dataset (including all ages, BMIs, and if they have children or not)


import os
print(os.getcwd())
```

```python
import csv

datareader = []
with open('Documents/Codecademy/medica_insurance_costs_project/python-portfolio-project-starter-files/python-portfolio-project-starter-files/insurance.csv', 'r', newline="") as file_to_read:
    data = csv.reader(file_to_read, dialect='excel')
    for i in data:
        datareader.append(i)
print(datareader)
```

```python
keys_columns = datareader[0]
values_rows = datareader[1:]
#print(keys_columns)
print(values_rows)
```

```python
datakeys_for_smoker_analysis = []
for data in keys_columns:
    if data == 'sex' or data == 'smoker' or data == 'region' or data == 'charges':
        datakeys_for_smoker_analysis.append(data)
    continue
datakeys_for_smoker_analysis
```

```python
datavalues_for_smoker_analysis = []
for data in values_rows:
    sublist = []
    sublist.append(data[1])
    sublist.append(data[4])
    sublist.append(data[5])
    sublist.append(data[6])
    datavalues_for_smoker_analysis.append(sublist)
print(datavalues_for_smoker_analysis, end='')
```

```python
sexes = []
smokers = []
regions = []
insurance_costs_string = []
for i in datavalues_for_smoker_analysis:
    sexes.append(i[0])
    smokers.append(i[1])
    regions.append(i[2])
    insurance_costs_string.append(i[3])
    
#smoker values transformed into int
smokers_int = []
for data in smokers:
    if data == 'no':
        data = 0
        smokers_int.append(data)
    elif data == 'yes':
        data = 1
        smokers_int.append(data)

print(smokers_int)
```

```python
#creating dictionary
sexes_smokers_zipped = list(zip(sexes, smokers_int))
def who_smokes_more():
    smokers_comparison = {}
    for key, value in zip(sexes, smokers_int):
        if key not in smokers_comparison:
            smokers_comparison[key] = []
        smokers_comparison[key].append(value)
    return smokers_comparison
print(who_smokes_more())
```

```python
def counting_smokers():
    men_who_smoke = 0
    women_who_smoke = 0
    for key, value in who_smokes_more().items():
        if key == 'female':
            women_who_smoke = sum(value)
        elif key == 'male':
            men_who_smoke = sum(value)
    return men_who_smoke, women_who_smoke

result1, result2 = counting_smokers()
print(result1, result2) #men in this dataset smoke more than women
```

```python
#Where are the majority of smoker men from?
#Where are the majority of smoker women from?
female_smokers_region = {'female': []}
male_smokers_region = {'male': []}
def smokers_regions():
    for i in datavalues_for_smoker_analysis:
        if 'female' in i and 'yes' in i:
            female_smokers_region['female'].append(i[1:])
        elif 'male' in i and 'yes' in i:
            male_smokers_region['male'].append(i[1:])
    return female_smokers_region, male_smokers_region

result3, result4 = smokers_regions()
print(result3, result4)
```

```python
#Where are the majority of smoker women from?
#A.: Southeast - 36 women
def female_smokers_per_region():
    northeast_female_smokers = 0
    northwest_female_smokers = 0
    southeast_female_smokers = 0
    southwest_female_smokers = 0
    for value in female_smokers_region['female']:
        if 'northeast' in value[1]:
            northeast_female_smokers +=1
        elif 'northwest' in value[1]:
            northwest_female_smokers +=1
        elif 'southeast' in value[1]:
            southeast_female_smokers +=1
        elif 'southwest' in value[1]:
            southwest_female_smokers +=1
    return f"The number of female smokers in Northeast is {northeast_female_smokers},\nThe number of female smokers in Northwest is {northwest_female_smokers},\nThe number of female smokers in Southeast is {southeast_female_smokers},\nThe number of female smokers in Southwest is {southwest_female_smokers}"

print(female_smokers_per_region())
```

```python
#Where are the majority of smoker men from?
#A.: Southeast - 55
def male_smokers_per_region():
    northeast_male_smokers = 0
    northwest_male_smokers = 0
    southeast_male_smokers = 0
    southwest_male_smokers = 0
    for value in male_smokers_region['male']:
        if 'northeast' in value[1]:
            northeast_male_smokers +=1
        elif 'northwest' in value[1]:
            northwest_male_smokers +=1
        elif 'southeast' in value[1]:
            southeast_male_smokers +=1
        elif 'southwest' in value[1]:
            southwest_male_smokers +=1
    return f"The number of male smokers in Northeast is {northeast_male_smokers},\nThe number of male smokers in Northwest is {northwest_male_smokers},\nThe number of male smokers in Southeast is {southeast_male_smokers},\nThe number of male smokers in Southwest is {southwest_male_smokers}"

print(male_smokers_per_region())
#Southeast region has the greater number of smokers in the dataset
```

```python
#BMI extraction
#BMI per sex
bmi = []
for data in values_rows:
    bmi.append(data[2])

#print(bmi)
zipped_bmi_per_sex = list(zip(sexes, bmi))
print(zipped_bmi_per_sex)
```

```python
def male_female_bmi_dict():
    bmi_per_sex_dict = {}
    for key, value in zip(sexes, bmi):
        if key not in bmi_per_sex_dict:
            bmi_per_sex_dict[key] = []
        bmi_per_sex_dict[key].append(value)
    return bmi_per_sex_dict
print(male_female_bmi_dict())
```

```python
#changing values from str to int
def str_to_int():
    converted_dict = {}
    for key, values in male_female_bmi_dict().items():
        converted_dict[key] = [float(value) for value in values]
    return converted_dict
result5 = str_to_int()
print(result5)
```

```python
#sum the values for each sex
def sum_bmi_values():
    sum_to_average_male = 0
    sum_to_average_female = 0
    for key, value in str_to_int().items():
        if key == 'female':
            sum_to_average_female = sum(value)
        else:
            sum_to_average_male = sum(value)
    return f"Sum of male BMI: {sum_to_average_male:.2f}\nSum of female BMI: {sum_to_average_female:.2f}"
result6 = sum_bmi_values()
print(result6)

#counting values of each key
def counting_values():
    count_female_bmi_values = 0
    count_male_bmi_values = 0
    for key, value in str_to_int().items():
        if key == 'female':
            count_female_bmi_values = len(value)
        count_male_bmi_values = len(value)
    return f"The number of female BMI values is {count_female_bmi_values}\nThe number of male BMI values is {count_male_bmi_values}"
result7 = counting_values()
print(result7)
```

```python
#average BMI values for each sex
def avg_bmi_values():
    avg_bmi_male = 20917.55/676*100
    avg_bmi_female = 20110.07/662*100
    return f"The average BMI for males in this dataset is {avg_bmi_male:.2f}%\nThe average BMI for females in this dataset is {avg_bmi_female:.2f}%"
result8 = avg_bmi_values()
print(result8)
print("These results of BMI for a human are nonsense")
```

```python
#median values for each sex
def calculate_median():
    median_dict = {}
    for key, value in str_to_int().items():
        sorted_values = sorted(value)  
        n = len(sorted_values)
        if n % 2 == 1:  # odd value, central value
            median = sorted_values[n // 2]
        else:  # even value, mean of the 2 central values
            median = (sorted_values[(n // 2) - 1] + sorted_values[n // 2]) / 2
        median_dict[key] = median
    return median_dict

median_result = calculate_median()
print(f"Median BMI values:\nMale: {median_result['male']:.2f}\nFemale: {median_result['female']:.2f}")
```

```python
#Insurance costs
zipped_sexes_insurance_cost = list(zip(sexes, insurance_costs_string))
#print(zipped_sexes_insurance_cost)

#create dict
def sexes_insurance_cost_dict():
    sexes_insurance_dict = {}
    for sex, values in zip(sexes, insurance_costs_string):
        if sex not in sexes_insurance_dict:
            sexes_insurance_dict[sex] = []
        sexes_insurance_dict[sex].append(values)
    return sexes_insurance_dict
result8 = sexes_insurance_cost_dict()
#print(result8)

#Turn string values into floats
def insurance_str_to_float():
    insu_str_to_float = {}
    for sex, values in sexes_insurance_cost_dict().items():
        insu_str_to_float[sex] = [float(num) for num in values]
    return insu_str_to_float
result9 = insurance_str_to_float()
print(result9)
```

```python
#calculate the average insurance cost for men and women
#median
def median_calculation_female_male():
    median_dictionary = {}
    for key, values in insurance_str_to_float().items():
        sort_values = sorted(values)
        n = len(sort_values)
        if n % 2 == 1:
            median = sort_values[n // 2]
        else:
            median = (sort_values[(n // 2) - 1] + sort_values[n // 2]) / 2
        median_dictionary[key] = median
    return median_dictionary
result10 = median_calculation_female_male()
print(f"Median Insurance cost values:\nMale: {result10['male']:.2f}\nFemale: {result10['female']:.2f}")
#The median Insurance cost for men in this dataset is 9369.62
#The median Insurance cost for women in this dataset is 9412.96
```
